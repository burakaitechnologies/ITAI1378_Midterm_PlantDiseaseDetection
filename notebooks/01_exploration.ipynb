```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from PIL import Image
from collections import Counter

# Set style
plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('Greens_d')
```

## 1. Load Dataset
Download the PlantVillage dataset from Kaggle:  
https://www.kaggle.com/datasets/emmarex/plantdisease

Place the extracted folder in `../data/plantvillage/`

```python
# Update this path to your dataset location
DATA_DIR = '../data/plantvillage/PlantVillage'

# List all classes
if os.path.exists(DATA_DIR):
    classes = sorted(os.listdir(DATA_DIR))
    print(f'Total classes: {len(classes)}')
    for i, cls in enumerate(classes):
        count = len(os.listdir(os.path.join(DATA_DIR, cls)))
        print(f'  {i+1}. {cls}: {count} images')
else:
    print(f'Dataset not found at {DATA_DIR}')
    print('Download from: https://www.kaggle.com/datasets/emmarex/plantdisease')
```

## 2. Class Distribution

```python
# Count images per class
if os.path.exists(DATA_DIR):
    class_counts = {}
    for cls in classes:
        class_counts[cls] = len(os.listdir(os.path.join(DATA_DIR, cls)))

    # Plot distribution
    fig, ax = plt.subplots(figsize=(14, 8))
    bars = ax.barh(list(class_counts.keys()), list(class_counts.values()), color='#4CAF50')
    ax.set_xlabel('Number of Images')
    ax.set_title('PlantVillage Dataset - Images per Class')
    plt.tight_layout()
    plt.show()

    print(f'\nTotal images: {sum(class_counts.values())}')
    print(f'Min class size: {min(class_counts.values())}')
    print(f'Max class size: {max(class_counts.values())}')
```

## 3. Sample Images

```python
# Display sample images from a few classes
if os.path.exists(DATA_DIR):
    sample_classes = classes[:6]  # First 6 classes
    fig, axes = plt.subplots(2, 3, figsize=(12, 8))

    for idx, cls in enumerate(sample_classes):
        ax = axes[idx // 3][idx % 3]
        cls_dir = os.path.join(DATA_DIR, cls)
        img_name = os.listdir(cls_dir)[0]
        img = Image.open(os.path.join(cls_dir, img_name))
        ax.imshow(img)
        ax.set_title(cls.replace('___', '\n'), fontsize=9)
        ax.axis('off')

    plt.suptitle('Sample Leaf Images', fontsize=14, fontweight='bold')
    plt.tight_layout()
    plt.show()
```

## 4. Image Properties

```python
# Check image dimensions and formats
if os.path.exists(DATA_DIR):
    widths, heights = [], []
    for cls in classes[:5]:  # Sample first 5 classes
        cls_dir = os.path.join(DATA_DIR, cls)
        for img_name in os.listdir(cls_dir)[:20]:  # 20 images per class
            img = Image.open(os.path.join(cls_dir, img_name))
            widths.append(img.size[0])
            heights.append(img.size[1])

    print(f'Width  - Min: {min(widths)}, Max: {max(widths)}, Mean: {np.mean(widths):.0f}')
    print(f'Height - Min: {min(heights)}, Max: {max(heights)}, Mean: {np.mean(heights):.0f}')
    print(f'\nAll images will be resized to 224x224 for ResNet50')
```
